```python
#Taking Data in and performing EDA

import pandas as pd
import pandas_profiling
data = pd.read_csv('data.csv', encoding = 'latin-1')
data = data[['tipo_prod','tipo_reg','marca_produto','modelo_produto','valor_venda','id_categoria']]
pandas_profiling.ProfileReport(data)
```

```python
#Establishing Baseline accuracy(The most frequent value/Total values x 100)

baseline_acc = (data['id_categoria'].value_counts().max())/int(data.shape[0])*100
print ("Baseline accuracy %",baseline_acc)
```

```python
#Label Binarizer for handling high cardinality categorical variables

from sklearn.preprocessing import LabelBinarizer

marca_lb = LabelBinarizer()
modelo_lb = LabelBinarizer()

X = marca_lb.fit_transform(data.marca_produto.values)
Xm = modelo_lb.fit_transform(data.modelo_produto.values)

dfOneHot = pd.DataFrame(X, columns = ["Marca_"+str(int(i)) for i in range(X.shape[1])])
data = pd.concat([data, dfOneHot], axis=1)

dfOneHot = pd.DataFrame(Xm, columns = ["Modelo_"+str(int(i)) for i in range(Xm.shape[1])])
data = pd.concat([data, dfOneHot], axis=1)

data = data.drop(['modelo_produto','marca_produto'], axis=1)
print (data.iloc[1])
```

```python
#Performing Train-Test split 75-25

from sklearn.model_selection import train_test_split
import numpy as np

data['is_train'] = np.random.uniform(0, 1, len(data)) <= .75
train, test = data[data['is_train']==True], data[data['is_train']==False]

print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:',len(test))

Y = data[['id_categoria']]
X = data.drop(['id_categoria'], axis=1)

features = X.columns[:]
target = Y.columns[:]


```

```python
#Training Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[features], train[target].values.ravel())
```

```python
#Evaluating Model

from sklearn.metrics import accuracy_score

preds = clf.predict(test[features])
print ("Accuracy On Test Dataset",accuracy_score(test['id_categoria'], preds))
```

```python
#Confusion Matrix 

pd.crosstab(test['id_categoria'], preds, rownames=['Actual Id'], colnames=['Predicted Id'])
```
